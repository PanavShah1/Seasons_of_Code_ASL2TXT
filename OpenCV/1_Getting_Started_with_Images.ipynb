## Splitting and merging color channels

```python
img_bgr = cv2.imread("scenery.jpg", 1)
b, g, r = cv2.split(img_bgr)

plt.figure(figsize=(20, 5))
plt.subplot(1, 4, 1)
plt.imshow(r, cmap="gray")
plt.title("Red channel")
plt.subplot(1, 4, 2)
plt.imshow(g, cmap="gray")
plt.title("Green channel")
plt.subplot(1, 4, 3)
plt.imshow(b, cmap="gray")
plt.title("Blue channel")
plt.subplot(1, 4, 4)
plt.imshow(img_bgr[:, :, ::-1])
plt.title("Merged Output")
```

## Converting to different Color Spaces

### Changing from BGR to RGB

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

### Changing to HSV color space

```python
img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(img_hsv)

plt.figure(figsize=(20, 5))
plt.subplot(1, 4, 1)
plt.imshow(h, cmap="gray")
plt.title("H channel")
plt.subplot(1, 4, 2)
plt.imshow(s, cmap="gray")
plt.title("S channel")
plt.subplot(1, 4, 3)
plt.imshow(v, cmap="gray")
plt.title("V channel")
plt.subplot(1, 4, 4)
plt.imshow(img_rgb)
plt.title("Merged Output")
```

## Modifying individual Channel

```python
h_new = h+10
img_merged = cv2.merge((h_new, s, v))
img_rgb = cv2.cvtColor(img_merged, cv2.COLOR_HSV2RGB)

plt.imshow(img_rgb)
```

## Saving Images

```python
cv2.imwrite("new_scenery.jpg", img_rgb)

Image(filename="new_scenery.jpg")
```
